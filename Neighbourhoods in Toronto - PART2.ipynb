## Explore Dataset and get Lattitude and Longitude


```python
import pandas as pd

#Using pandas to get the table from wikipedia webpage
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

df_nb=pd.read_html(url, header=0)[0]
```

```python
#Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.
df_nb = df_nb[df_nb.Borough != 'Not assigned']
df_nb = df_nb.rename(columns={'Postal Code': 'PostalCode'})

#If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.
for index, row in df_nb.iterrows():
    if row['Neighbourhood'] == 'Not assigned':
        row['Neighbourhood'] = row['Borough']

df_nb = df_nb.rename(columns={'Neighbourhood': 'Neighborhood'})

df_nb = df_nb.reset_index(drop=True)
df_nb.head(12)
```

### Import location coordinates from Geospatial_data

```python
import pandas as pd

# read csv file from the address
Geospacial_Coordinates = pd.read_csv('http://cocl.us/Geospatial_data', sep = ',')
Geospacial_Coordinates = Geospacial_Coordinates.rename(columns={'Postal Code': 'PostalCode'})

# examine the original input data
Geospacial_Coordinates.head()
```

```python
# concatinate the original data from wikipedia with Geospacial_Coordinates csv to get the Latitude and Longitude
neighborhoods = pd.merge(df_nb, Geospacial_Coordinates, on='PostalCode')

#df_fnb = neighborhoods.drop(['Postal Code'], axis = 1)
neighborhoods.head(12)
```
